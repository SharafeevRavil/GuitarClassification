```python
with open('F:\Repos\GuitarClassification\Desktop\MainApplication\AIModule\YOLO\\rofl\_annotations.txt', 'r') as anno:
    line = anno.readline()
    recs = line.split()[1:]
    image = Image.open(line.split()[0])

    fig, ax = plt.subplots()
    ax.imshow(image)
    for rec in recs:
        coords = rec.split(',')
        
        rect = patches.Rectangle((int(coords[0]), int(coords[1])), int(coords[2])-int(coords[0]), int(coords[3])-int(coords[1])-1, linewidth=1, edgecolor=numpy.random.rand(3,), facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```
