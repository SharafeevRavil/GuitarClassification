```python
file = open('./keras-yolo3-master/model_data/voc_classes.txt',mode='r')
all_classes = file.read()
file.close()
classes = all_classes.split('\n')

with open('.\\rofl\_annotations.txt', 'r') as anno:
    line = anno.readline()
    recs = line.split()[1:]
    image = Image.open(line.split()[0])

    fig, ax = plt.subplots()
    ax.imshow(image)
    for rec in recs:
        coords = rec.split(',')
        
        color = numpy.random.rand(3,)
        x = int(coords[0])
        y = int(coords[1])
        xl = int(coords[2])
        yl = int(coords[3])
        yd = numpy.random.randint(0, 20)
        rect = patches.Rectangle((x, y+yd), xl-x, yl-y-1-2*yd, linewidth=1, edgecolor=color, facecolor='none')
        ax.text(int(coords[0]), numpy.random.randint(0, 192), classes[int(coords[4])], color = color, size='large')
        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```
