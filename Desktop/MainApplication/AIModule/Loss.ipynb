```python
from matplotlib import pyplot
import json
```

```python
loss_file = open('./data/output/loss.json')
acc_file = open('./data/output/avg_acc.json')
loss = json.load(loss_file)
avg_acc = json.load(acc_file)
```

```python
pyplot.title('Loss by batches')
pyplot.plot(loss, label='loss')
pyplot.legend()
pyplot.show()
```

```python
pyplot.title('Accuracy by batches')
pyplot.plot(avg_acc, label='avg_acc')
pyplot.legend()
pyplot.show()
```
